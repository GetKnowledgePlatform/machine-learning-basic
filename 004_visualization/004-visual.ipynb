# Plot

![Normal](randn.svg)

```python
x = np.linspace(0, 10, 500)
```

```python
rng = np.random.RandomState(0)
range_data = rng.randn(500, 6)
#range_data
```

```python
y = np.cumsum(range_data, axis=0)
```

```python
plt.plot(x, y)
plt.legend('ABCDEF', ncol=2, loc='upper left');
```

# Histogram

```python
data = np.random.multivariate_normal([0, 0], [[5, 2], [2, 2]], size=2000)
data = pd.DataFrame(data, columns=['x', 'y'])


plt.hist(data['x'], alpha=0.5, bins=50)
plt.hist(data['y'], alpha=0.5, bins=50);
```

```python
sns.displot(data['x'], kind='kde')
sns.displot(data['y'], kind='kde');
```

# Pairplot

```python
iris = sns.load_dataset("iris")
iris.head()
```

```python
sns.pairplot(iris, hue='species', size=2.5);
```

```python
tips = sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']

grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

```python
sns.boxplot( y=tips["total_bill"], x=tips["day"] );
```

![BoxPlot](boxplot.png)

```python
sns.jointplot(x="total_bill", y="tip", data=tips, kind='reg');
```
